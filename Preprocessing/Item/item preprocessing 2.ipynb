```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import os

data = pd.read_csv('games.csv')
data.head()
```

```python
data1 = data.drop(['About the game','Reviews','Header image','Website','Support url','Support email','Metacritic url','Notes','Screenshots'], axis = 'columns')
data1.head()
```

```python
#결측치(Missing data)
print('전체 데이터 건수:', len(data))
```

```python
# Name 데이터는 object값을 가지므로 인코딩을 진행해보았다.

from sklearn.preprocessing import LabelEncoder

Name = data1['Name']

df_Name = pd.DataFrame(Name)

label_encoder = LabelEncoder()
df_Name['Category_encoded_Name'] = label_encoder.fit_transform(df_Name['Name'])

print("Original DataFrame:")
print(df_Name[['Name']])
print("\nDataFrame with Label Encoded Column:")
print(df_Name[['Category_encoded_Name']])
```

```python
# Required age가 어느정도로 분포되어있는지에 대해 알기 위해 중복된 데이터를 삭제한 후 오름차순으로 정렬하였다.

from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler
import pandas as pd

Required_age = data1['Required age']

df_Required_age = pd.DataFrame(Required_age)

df_Required_age_no_duplicates = df_Required_age.drop_duplicates()

df_Required_age_sorted = df_Required_age_no_duplicates.sort_values(by='Required age', ascending=True)


print("\nDataFrame with Duplicates Removed:")
print(df_Required_age_no_duplicates)

print("\nDataFrame Sorted by Requiredage in Ascending Order:")
print(df_Required_age_sorted)

# 최대 1세부터 21세까지 나와있고 0은 전체 이용가이다.

# Min-Max Scaling (0과 1 사이의 범위로 스케일링)
min_max_scaler = MinMaxScaler()
min_max_scaled_data = min_max_scaler.fit_transform(df_Required_age)

print("Min-Max Scaled Data:")
print(min_max_scaled_data)

# Standard Scaling (Z-score normalization)
standard_scaler = StandardScaler()
standard_scaled_data = standard_scaler.fit_transform(df_Required_age)

print("\nStandard Scaled Data:")
print(standard_scaled_data)

# Robust Scaling (중앙값과 IQR을 사용한 스케일링)
robust_scaler = RobustScaler()
robust_scaled_data = robust_scaler.fit_transform(df_Required_age)

print("\nRobust Scaled Data:")
print(robust_scaled_data)
```

```python
# DLC count가 어느정도로 분포되어있는지에 대해 알기 위해 중복된 데이터를 삭제한 후 오름차순으로 정렬하였다.

DLC_count = data1['DLC count']

df_DLC_count = pd.DataFrame(DLC_count)

df_DLC_count_no_duplicates = df_DLC_count.drop_duplicates()

df_DLC_count_sorted = df_DLC_count_no_duplicates.sort_values(by='DLC count', ascending=True)


print("\nDataFrame with Duplicates Removed:")
print(df_DLC_count_no_duplicates)

print("\nDataFrame Sorted by Requiredage in Ascending Order:")
print(df_DLC_count_sorted)

# Min-Max Scaling (0과 1 사이의 범위로 스케일링)
min_max_scaler = MinMaxScaler()
min_max_scaled_data = min_max_scaler.fit_transform(df_DLC_count)

print("Min-Max Scaled Data:")
print(min_max_scaled_data)

# Standard Scaling (Z-score normalization)
standard_scaler = StandardScaler()
standard_scaled_data = standard_scaler.fit_transform(df_DLC_count)

print("\nStandard Scaled Data:")
print(standard_scaled_data)

# Robust Scaling (중앙값과 IQR을 사용한 스케일링)
robust_scaler = RobustScaler()
robust_scaled_data = robust_scaler.fit_transform(df_DLC_count)

print("\nRobust Scaled Data:")
print(robust_scaled_data)
```

```python
# Full audio languages  데이터는 object값을 가지므로 인코딩을 진행해보았다.

from sklearn.preprocessing import LabelEncoder

Full_audio_languages  = data1['Full audio languages']

df_Full_audio_languages  = pd.DataFrame(Full_audio_languages)

label_encoder = LabelEncoder()
df_Full_audio_languages ['Category_encoded_Full_audio_languages'] = label_encoder.fit_transform(df_Full_audio_languages['Full audio languages'])

print("Original DataFrame:")
print(df_Full_audio_languages [['Full audio languages']])
print("\nDataFrame with Label Encoded Column:")
print(df_Full_audio_languages [['Category_encoded_Full_audio_languages']])


```

```python
# MAC 칼럼 데이터는 TRUE or False 값을 갖는 bool값이다. 따라서 Mac사용 유저에게 추천해줄 수 있는 데이터를 추출한다.

Mac = data1['Mac']

df_Mac = pd.DataFrame(Mac)

filtered_Mac = df_Mac[df_Mac['Mac'] == True]

print("\nFiltered DataFrame:")
print(filtered_Mac)

# 76983개의 게임 데이터 중 15156 개의 게임 데이터만 Mac 유저에게 추천이 가능하다.
```

```python
# Metacritic score 칼럼 데이터는 비평가들에게 받은 점수로 대부분의 데이터가 0으로 나와있기 떄문에 먼저 0값을 갖는 데이터를 삭제해준다.
#칼럼 데이터는 TRUE or False 값을 갖는 bool값이다. 따라서 Mac사용 유저에게 추천해줄 수 있는 데이터를 추출한다.

Metacritic_score = data1['Metacritic score']

df_Metacritic_score = pd.DataFrame(Metacritic_score)

df_Metacritic_score

filtered_Metacritic_score = df_Metacritic_score[(df_Metacritic_score != 0).all(axis=1)]

print("\nFiltered DataFrame:")
print(filtered_Metacritic_score)

mean_Metacritic_score = filtered_Metacritic_score.mean()
max_Metacritic_score = filtered_Metacritic_score.max()
min_Metacritic_score = filtered_Metacritic_score.min()
median_Metacritic_score = filtered_Metacritic_score.median()
std_Metacritic_score = filtered_Metacritic_score.std()

print("---------reslut---------")
print("Mean Values:")
print(mean_Metacritic_score)

print("Max Values:")
print(max_Metacritic_score)

print("Min Values:")
print(min_Metacritic_score)

print("Median Values:")
print(median_Metacritic_score)

print("Standard Deviation Values:")
print(std_Metacritic_score)

# 76983개의 게임 데이터 중 3792 개의 게임 데이터의 평균, 최대, 최소, 중간 값, 표준편차를 구했다.
```

```python
# Positive 칼럼 데이터가 어떤 것을 의미하는지 몰라 Metacritic score와 비슷하게 분석을 진행해보았다.

Positive = data1['Positive']

df_Positive = pd.DataFrame(Positive)

df_Positive

filtered_Positive = df_Positive[(df_Positive != 0).all(axis=1)]

print("\nFiltered DataFrame:")
print(filtered_Positive)

mean_Positive = filtered_Positive.mean()
max_Positive = filtered_Positive.max()
min_Positive = filtered_Positive.min()
median_Positive = filtered_Positive.median()
std_Positive = filtered_Positive.std()

print("---------reslut---------")
print("Mean Values:")
print(mean_Positive)

print("Max Values:")
print(max_Positive)

print("Min Values:")
print(min_Positive)

print("Median Values:")
print(median_Positive)

print("Standard Deviation Values:")
print(std_Positive)

# 76983개의 게임 데이터 중 57539개의 게임 데이터의 평균, 최대, 최소, 중간 값, 표준편차를 구했다.
# 평균 값이 1392인데 중앙 값이 19인 것을 미루어 보아 소수의 데이터가 굉장히 큰 값을 가지고 있는 왼쪽으로 꼬리가 긴 데이터를 가지고 있는 것을 알 수 있다.
```

```python
# Score rank 칼럼 데이터는 42개의 데이터 만이 non-null 데이터 이므로
# 칼럼을 삭제하는 것이 좋을 것 같다.
```

```python
# Recommendations 값이 0이라면 사실 상 해당 게임은 유저들이 전혀 하지 않는 게임이라고 할 수 있다.
# 따라서 이 데이터들을 우선적으로 삭제하려고 한다.

Recommendations = data1['Recommendations']

df_Recommendations = pd.DataFrame(Recommendations)

df_Recommendations

filtered_Recommendations = df_Recommendations[(df_Recommendations != 0).all(axis=1)]

print("\nFiltered DataFrame:")
print(filtered_Recommendations)

# 총 13057개의 데이터가 추천되었음을 알 수 있다.
# 이 데이터를 내림차순으로 정리하면 다음과 같다.

df_sorted = df_Recommendations.sort_values(by='Recommendations', ascending=False)

print("\nDataFrame Sorted by Recommendations in Descennding Order:")
print(df_sorted)

filtered_Recommendations = df_Recommendations[(df_Recommendations != 0).all(axis=1)]

mean_Recommendations = filtered_Recommendations.mean()
median_Recommendations= filtered_Recommendations.median()

print("Mean Values:")
print(mean_Recommendations)
print("Median Values:")
print(median_Recommendations)

# 평균 값이 4946인데 중앙 값이 410인 것을 미루어 보아 소수의 데이터가 굉장히 큰 값을 가지고 있는 왼쪽으로 꼬리가 긴 데이터를 가지고 있는 것을 알 수 있다.
```

```python
# Average playtime two weeks 값이 0이라면 사실 상 해당 게임은 유저들이 전혀 하지 않는 게임이라고 할 수 있다.
# 따라서 이 데이터들을 우선적으로 삭제하려고 한다.

Average_playtime_two_weeks = data1['Average playtime two weeks']

df_Average_playtime_two_weeks = pd.DataFrame(Average_playtime_two_weeks)

df_Average_playtime_two_weeks

filtered_Average_playtime_two_weeks = df_Average_playtime_two_weeks[(df_Average_playtime_two_weeks != 0).all(axis=1)]

print("\nFiltered DataFrame:")
print(filtered_Average_playtime_two_weeks)
```
